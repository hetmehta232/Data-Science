```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#DataFrames:

m = np.random.randint(0,100,(5,5))
print(m)
table = pd.DataFrame(m,['sakshi','het','vrund','dweeti','harshit'],['python','dsa','dbms','ooad','stb'])
```

```python
#Perticular column:

table['python']
```

```python
table
```

```python
x = np.linspace(0,5,11)
y =  x**2
```

```python
plt.plot(x,y,'r--')
plt.plot(x,y,'k*')
# line style: --,-,-.,:
plt.plot(y,x,'b:')
plt.plot(y,x,'g^')
#line colour: red(r),blue(b),green(g),Yellow(y),Magenta(m),Black(k),Cyan(c)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title("Title")
plt.xlim(0,25)
plt.ylim(-5,30)
```

```python
name=table.index
plt.plot(name,table['python'],'r--')
plt.plot(name,table['dsa'],'g--')
plt.xlim(0,5)
plt.ylim(-5,100)
```

```python
#Subplot:

plt.subplot(1,3,1)
plt.plot(name,table['python'],'r--')

plt.subplot(1,3,2)
plt.plot(name,table['dsa'],'c-.')

plt.subplot(1,3,3)
plt.plot(name,table['stb'],'ko-')
```

```python
plt.subplot(2,2,1)
plt.plot(name,table['python'],'r--',label='Python')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title("| Python |")
plt.legend(loc=8)

plt.subplot(2,2,2)
plt.plot(name,table['dsa'],'c-.',label='DSA')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title("| DSA |")
plt.legend(loc=1)


plt.subplot(2,2,3)
plt.plot(name,table['stb'],'ko-',label='STB')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title("| Software Testing |")
plt.legend(loc=5)


plt.subplot(2,2,4)
plt.plot(name,table['ooad'],'m-',label='OOAD')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title("| OOAD |")

plt.tight_layout()
plt.legend(loc=8)
```

```python
x = np.linspace(0,5,11)
y =  x**2

fig=plt.figure()

axes=fig.add_axes([0.1,0.1,0.8,0.85])
axes.plot(x,y,'b^-',label='PNJ')


axes=fig.add_axes([0.17,0.48,0.4,0.42])

axes.plot(y,x,'r-*',label='Leo')
axes.set_title('Future')
axes.set_xlabel('X label')
axes.set_ylabel('Y label')

axes.legend(loc=4)
```

```python
fig1,axes=plt.subplots(2,2,layout='constrained')
axes[1,1].plot(name,table['ooad'],'r-',label='OOAD',alpha=0.5)
axes[1,1].set_xlabel('X')
axes[1,1].set_ylabel('Y')
axes[1,1].legend()
axes[1,1].grid(True)


axes[0,0].plot(name,table['python'],'b-',label='PYTHON',alpha=0.5)
axes[0,0].set_xlabel('X axis')
axes[0,0].set_ylabel('Y axis')
axes[0,0].legend()
axes[0,0].grid(True)


axes[0,1].plot(name,table['stb'],'g-',label='Software Testing',alpha=0.5)
axes[0,1].set_xlabel('X axis')
axes[0,1].set_ylabel('Y axis')
axes[0,1].legend()
axes[0,1].grid(True)


axes[1,0].plot(name,table['dbms'],'k-',label='DBMS',alpha=0.5)
axes[1,0].set_xlabel('X axis')
axes[1,0].set_ylabel('Y axis')
axes[1,0].legend()
axes[1,0].grid(True)
```

```python
#Univariant    #Multivariant
#Pie           #Scatter
#Histogram     #Barchart
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Pie,Scatter,Stairs,Stackplot,Stem,Line,Hist2d,Hexbin:
x=np.random.normal(0,10,200)
y=np.random.normal(0,20,200)
c=np.random.normal(0,20,200)
s=np.random.normal(0,10,200)

plt.scatter(x,y,s=100,c=c,marker='o',alpha=1,cmap='cool',linewidth=1,edgecolor='k')
plt.colorbar()
```

```python
plt.bar(x,y)
```

```python
plt.scatter(x,y,s=100,c=c,marker='p',alpha=1,cmap='cool',linewidth=1,edgecolor='k')
plt.colorbar()
```

```python
plt.hist(x,bins=10)
plt.hist(y,bins=50)
plt.hist(s,bins=70)
plt.hist(c,bins=100)
```

```python
plt.hist2d(x,y,bins=10,cmap='Reds')
```

```python
x=np.random.normal(0,10,25)
y=np.random.normal(0,20,25)
plt.hexbin(x,y,bins=100,cmap='Blues',gridsize=10)
```

```python
plt.stem(x,y)
```
